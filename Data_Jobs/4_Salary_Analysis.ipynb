3. How well do jobs and skills pay for Data Analyst? 

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
#Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#Limit data to Malaysia
df_my = df[df['job_country']=='Malaysia']
```

```python
#drop rows with no salary
df_my = df_my.dropna(subset=['salary_year_avg'])
df_my
```

Salary Distribution of Top 5 Data Jobs in Malaysia

```python
top5_job_titles = df_my['job_title_short'].value_counts().index[:5].tolist()

#Filter data to top 5 jobs
df_top5_my = df_my[df_my['job_title_short'].isin(top5_job_titles)]
df_top5_my
```

```python
#get order of job by salary median
job_order = df_top5_my.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
job_order
```

```python
sns.boxplot(data=df_top5_my, x='salary_year_avg', y='job_title_short', order=job_order)
sns.set_theme(style='ticks')
sns.despine()

# this is all the same
plt.title('Salary Distributions of Data Jobs in Malaysia')
plt.xlabel('Yearly Salary (USD)')
plt.ylabel('')
plt.xlim(0, 200000) 
ticks_x = plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K')
plt.gca().xaxis.set_major_formatter(ticks_x)
```

Expand scope to include Singapore data jobs for larger sample size

```python
#Limit data to Malaysia and Singapore
df_my_sg = df[(df['job_country']=='Malaysia') | (df['job_country']=='Singapore')]
df_my_sg = df_my_sg.dropna(subset=['salary_year_avg'])
```

```python
top5_job_titles = df_my_sg['job_title_short'].value_counts().index[:5].tolist()

#Filter data to top 5 jobs
df_top5_my_sg = df_my_sg[df_my_sg['job_title_short'].isin(top5_job_titles)]
df_top5_my_sg
```

```python
#get order of job by salary median
job_order_my_sg = df_top5_my_sg.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
job_order_my_sg
```

```python
sns.boxplot(data=df_top5_my_sg, x='salary_year_avg', y='job_title_short', order=job_order_my_sg)
sns.set_theme(style='ticks')
sns.despine()

plt.title('Salary Distributions of Data Jobs in Malaysia & Singapore')
plt.xlabel('Yearly Salary (USD)')
plt.ylabel('')
plt.xlim(0, 250000) 
ticks_x = plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K')
plt.gca().xaxis.set_major_formatter(ticks_x)
```

Highest Paid Skills and Median Pay of Most Demanded Skills

```python
df_da_my = df_my[df_my['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_da_my_explode = df_da_my.explode('job_skills')
df_da_my_explode.head(5)
# df_da_my_explode[['salary_year_avg', 'job_skills']].head(5)
```

```python
df_da_my_top_pay = df_da_my_explode.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='median', ascending=False)

#get top 10 paying job skills
df_da_my_top_pay = df_da_my_top_pay.head(10)
df_da_my_top_pay
```

```python
df_da_my_skills = df_da_my_explode.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False)

#get top 10 demanded job skills (by count of job postings)
df_da_my_skills = df_da_my_skills.head(10).sort_values(by='median', ascending=False)
df_da_my_skills
```

```python
fig, ax = plt.subplots(2, 1)  

sns.set_theme(style='ticks')

# Top 10 Highest Paid Skills for Data Analysts
sns.barplot(data=df_da_my_top_pay, x='median', y=df_da_my_top_pay.index, hue='median', ax=ax[0], palette='dark:b_r')
ax[0].legend().remove()
ax[0].set_title('Highest Paid Skills for Data Analysts in Malaysia')
ax[0].set_ylabel('')
ax[0].set_xlabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
ax[0].set_xlim(0, 120000)

# Top 10 Most Demanded Skills for Data Analysts')
sns.barplot(data=df_da_my_skills, x='median', y=df_da_my_skills.index, hue='median', ax=ax[1], palette='light:b')
ax[1].legend().remove()
ax[1].set_title('Most Demanded Skills for Data Analysts in Malaysia')
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary (USD)')
ax[1].set_xlim(ax[0].get_xlim())  # Set the same x-axis limits as the first plot
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

plt.tight_layout()
```
